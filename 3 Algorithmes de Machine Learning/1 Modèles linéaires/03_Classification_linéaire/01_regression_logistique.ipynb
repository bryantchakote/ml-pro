# La regression logistique

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Visualisation de la fonction logistique en 2D

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Générer des valeurs x de -10 à 10
x = np.linspace(-10, 10, 200)

# Calculer les valeurs sigmoïdes pour chaque x
y = sigmoid(x)

# Créer le graphique
plt.figure(figsize=(10, 6))
plt.plot(x, y, color="blue", label="Sigmoïde")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

```

On peut la visualiser en 3D pour voir la surface de décision. En fait, les points de la classe 1 sont au dessus de la surface et ceux de la classe 0 en dessous.

```python

# Définition de la fonction sigmoïde
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Création d'une grille pour Xtheta
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = sigmoid(X)  # Nous utilisons uniquement X pour illustrer l'effet

# Configuration du graphique 3D
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# Surface sigmoïde
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

ax.set_title('Visualisation 3D de la fonction sigmoïde\n')
# point de vue
ax.view_init(elev=30, azim=120)
plt.show()
```
