# Encodage

```python
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder

# Pas recommandé en pratique, ne marche pas bien avec les matrices creuses
from sklearn import set_config

set_config(transform_output="pandas")
```

---

#### **1. Tips**

```python
df_1 = sns.load_dataset("tips")
df_1.head()  # target: tip
```

```python
display(
    df_1["sex"].unique(),
    df_1["smoker"].unique(),
    df_1["day"].unique(),
    df_1["time"].unique(),
)
```

```python
cols_1 = ["sex", "smoker", "day", "time"]
encoder_1 = OneHotEncoder(drop="first", sparse_output=False)
encoder_1.fit_transform(df_1[cols_1])
```

#### **2. Penguins**

```python
df_2 = sns.load_dataset("penguins")
df_2.dropna(inplace=True)
df_2.head()  # target: species
```

```python
display(
    df_2["species"].unique(),
    df_2["island"].unique(),
    df_2["sex"].unique(),
)
```

```python
cols_2 = ["island", "sex"]
label_2 = "species"

cols_encoder_2 = OneHotEncoder(drop="first", sparse_output=False)
cols_encoder_2.fit(df_2[cols_2])

label_encoder_2 = LabelEncoder()  # ou LabelBinarizer en fonction du modèle à entraîner
label_encoder_2.fit(df_2[label_2])

display(
    cols_encoder_2.transform(df_2[cols_2]),
    label_encoder_2.transform(df_2[label_2]),
)
```

#### **3. Flights**

```python
df_3 = sns.load_dataset("flights")
df_3.head()  # target: passengers
```

```python
col_3 = "month"
categories_3 = df_3[col_3].cat.categories

encoder_3 = OrdinalEncoder(categories=[categories_3])
encoder_3.fit_transform(df_3[[col_3]])

# Possible en OneHot, mais cela aurait créé 11 colonnes (avec drop="first").
# En pratique, la meilleure chose à faire est de tester différentes approches.
```

#### **4. Exercise**

```python
df_4 = sns.load_dataset("exercise")
df_4.head()  # target: pulse
```

```python
display(
    df_4["diet"].unique(),
    df_4["time"].unique(),
    df_4["kind"].unique(),
)
```

```python
cols_4 = ["diet", "time", "kind"]
categories_4 = [df_4[col].cat.categories for col in cols_4]

encoder_4 = OrdinalEncoder(categories=categories_4)
encoder_4.fit(df_4[cols_4])
encoder_4.transform(df_4[cols_4])
```

```python
# Il vaut mieux extraire la valeur numérique dans des variables comme time.
# L'écart aurait été différent entre les valeurs qu'on aurait perdu l'info.
df_4["time"].str.extract(r"(\d+)").astype(int)
```

#### **5. Taxis**

```python
df_5 = sns.load_dataset("taxis")
df_5.dropna(inplace=True)
df_5.head()  # target: total_bill
```

```python
display(
    df_5["color"].unique(),
    df_5["payment"].unique(),
    df_5["pickup_zone"].nunique(),  # beaucoup de categories
    df_5["dropoff_zone"].nunique(),  # beaucoup de categories
    df_5["pickup_borough"].unique(),
    df_5["dropoff_borough"].unique(),
)
```

a. `pickup_zone` et `dropoff_zone` contiennent beaucoup trop de catégories.
- Afin de remédier à cela, les possibilités qui s'offrent à nous sont :
  - Utiliser un target encoding (à voir prochainement).
  - Utiliser les coordonnées GPS des lieux plutôt que leurs noms.
  - Mettre de côté ces variables et se servir uniquement des `pickup_borough` et `dropoff_borough` (ou d'autres clusters construits par feature engineering).
- Toujours est-il qu'il faut expérimenter et de garder la technique qui donne de meilleurs résultats. 

b. `pickup_borough` et `dropoff_borough` ont les mêmes valeurs, à l'exception de "Staten Island" qui se retrouve dans la 2e variable mais pas dans la 1ère.
- Seulement 2 observations sur 6341 sont concernées, nous allons les supprimer à cause du manque de représentativité.

```python
df_5_new = df_5.query("dropoff_borough != 'Staten Island'").drop(
    ["pickup_zone", "dropoff_zone"], axis=1
)

cols_5_1 = ["color", "payment"]
cols_5_2 = ["pickup_borough", "dropoff_borough"]
categories_5_2 = df_5["dropoff_borough"].unique()

encoder_5_1 = OneHotEncoder(drop="first", sparse_output=False)
encoder_5_2 = OneHotEncoder(
    categories=[categories_5_2, categories_5_2],  # répétition pour pickup et dropoff
    drop="first",
    sparse_output=False,
)

display(
    encoder_5_1.fit_transform(df_5_new[cols_5_1]),
    encoder_5_2.fit_transform(df_5_new[cols_5_2]),
)
```
